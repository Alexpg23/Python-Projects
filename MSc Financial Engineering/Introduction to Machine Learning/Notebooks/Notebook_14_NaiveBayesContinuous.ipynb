# Naive Bayes Continuous Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')

%matplotlib inline
```

## Loading the Iris Dataset

```python
from sklearn.datasets import load_iris
import pandas as pd
    
iris = load_iris()
X = pd.DataFrame(iris.data)
X.columns = ['sepalLength','sepalWidth','petalLength','petalWidth']
X.head()
```

```python
y = iris.target
```

## Visualising the Decision Space

```python
from sklearn.naive_bayes import GaussianNB
```

```python
featureX = 'sepalLength'
featureY = 'sepalWidth'
```

```python
X_2 = X[[featureX,featureY]]
# I only want a binary classification to start with
y_2 = (np.array(y)==2).astype(np.int)
```

```python
clf = GaussianNB()
clf.fit(X_2, y_2)

prediction = clf.predict(X_2)
```

```python
import HelpfulFunctions as hp
```

```python
plt.figure(figsize=(8,8))

hp.featureSpacePlot(featureX,featureY,X,y_2,clf,plt);
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_2, prediction)
```
