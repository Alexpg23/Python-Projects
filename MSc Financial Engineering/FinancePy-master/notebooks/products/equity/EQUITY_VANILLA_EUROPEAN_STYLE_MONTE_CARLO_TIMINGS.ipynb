# EQUITY VANILLA OPTION USING MONTE CARLO PLUS TIMINGS

Valuation and Risk of a simple Vanilla Equity Option

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

```python
from financepy.utils import *
from financepy.products.equity import EquityVanillaOption
from financepy.market.curves.discount_curve_flat import *
from financepy.models.black_scholes import *
```

# Define the Call and Put Options

```python
valuation_date = Date(1, 4, 2016)
```

```python
expiry_date = Date(1, 6, 2018)
```

```python
expiry_date
```

```python
strike_price = 100.0
```

```python
call_option = EquityVanillaOption(expiry_date, strike_price, OptionTypes.EUROPEAN_CALL)
```

```python
stock_price = 100.0
volatility = 0.20
interest_rate = 0.05
dividend_yield = 0.03
```

```python
discount_curve = DiscountCurveFlat(valuation_date, interest_rate)
```

```python
dividend_curve = DiscountCurveFlat(valuation_date, dividend_yield)
```

```python
model = BlackScholes(volatility)
```

```python
v = call_option.value(valuation_date, stock_price, discount_curve, dividend_curve, model)
print(v)
```

```python
v_exact = [v] * 10 
```

# Timing Analysis

```python
seed = 141
```

```python
num_paths_list = np.arange(1,11,1) * 10000
```

## No Numpy Vectorisation and No NUMBA

```python
v_mc_NO_NUMPY = []
timings_NO_NUMPY = []

for num_paths in num_paths_list:
    start = time()
    v = call_option.value_mc_nonumba_nonumpy(valuation_date, stock_price, discount_curve, 
                           dividend_curve, model, num_paths, seed, useSobol=False)
    end = time()
    elapsed = end - start
    timings_NO_NUMPY.append(elapsed)
    v_mc_NO_NUMPY.append(v)
```

```python
plt.plot(num_paths_list, timings_NO_NUMPY, 'o-')
plt.xlabel("NumPaths")
plt.ylabel("CPU Wall Time (s)")
plt.title("NUMPY ONLY");
```

## Standard Monte Carlo using Numpy

```python
num_paths_list = np.arange(1,11,1) * 1000000
```

```python
v_mc_NUMPY_ONLY = []
timings_NUMPY_ONLY = []

for num_paths in num_paths_list:
    start = time()
    v = call_option.value_mc_numpy_only(valuation_date, stock_price, discount_curve, 
                           dividend_curve, model, num_paths, seed, useSobol=False)
    end = time()
    elapsed = end - start
    timings_NUMPY_ONLY.append(elapsed)
    v_mc_NUMPY_ONLY.append(v)
```

```python
plt.plot(num_paths_list, timings_NUMPY_ONLY, 'o-')
plt.xlabel("NumPaths")
plt.ylabel("CPU Wall Time (s)")
plt.title("NUMPY ONLY");
```

```python
plt.title("NUMPY ONLY")
plt.plot(num_paths_list, v_mc_NUMPY_ONLY, label="MC")
plt.plot(num_paths_list, v_exact, label="Exact")
plt.xlabel("Num Paths")
plt.ylabel("Option Price")
plt.legend();
```

## Standard Monte Carlo using Numpy and Numba

```python
v_mc_NUMPY_NUMBA = []
timings_NUMPY_NUMBA = []

for num_paths in num_paths_list:
    start = time()
    v = call_option.value_mc_numba_only(valuation_date, stock_price, discount_curve, 
                           dividend_curve, model, num_paths, seed)
    end = time()
    elapsed = end - start
    timings_NUMPY_NUMBA.append(elapsed)
    v_mc_NUMPY_NUMBA.append(v)
```

```python
plt.plot(num_paths_list, timings_NUMPY_NUMBA, 'o-')
plt.xlabel("NumPaths")
plt.ylabel("CPU Wall Time (s)")
plt.title("NUMPY NUMBA");
```

```python
plt.title("NUMPY NUMBA")
plt.plot(num_paths_list, v_mc_NUMPY_NUMBA, label="MC NUMPY NUMBA")
plt.plot(num_paths_list, v_exact, label="Exact")
plt.xlabel("Num Paths")
plt.ylabel("Option Price")
plt.legend();
```

## Using Sobol Quasi-Random Numbers

```python
v_mc_SOBOL_NUMPY = []
timings_SOBOL_NUMPY = []

for num_paths in num_paths_list:
    start = time()
    v = call_option.value_mc(valuation_date, stock_price, discount_curve, 
                           dividend_curve, model, num_paths, seed, useSobol=True)
    end = time()
    elapsed = end - start
    timings_SOBOL_NUMPY.append(elapsed)
    v_mc_SOBOL_NUMPY.append(v)
```

```python
plt.plot(num_paths_list, timings_SOBOL_NUMPY, 'o-')
plt.xlabel("NumPaths")
plt.ylabel("CPU Wall Time (s)")
plt.title("NUMPY SOBOL");
```

```python
plt.title("NUMPY SOBOL")
plt.plot(num_paths_list, v_mc_SOBOL_NUMPY, label="NUMPY_SOBOL")
plt.plot(num_paths_list, v_exact, label="EXACT")

plt.xlabel("Num Paths")
plt.ylabel("Option Price")
plt.legend()
```

```python
plt.title("NUMPY SOBOL")
plt.plot(num_paths_list, v_mc_SOBOL_NUMPY, label="NUMPY_SOBOL")
plt.plot(num_paths_list, v_exact, label="EXACT")
plt.plot(num_paths_list, v_mc_NUMPY_ONLY, label="NUMPY_ONLY")

plt.xlabel("Num Paths")
plt.ylabel("Option Price")
plt.legend()
```
