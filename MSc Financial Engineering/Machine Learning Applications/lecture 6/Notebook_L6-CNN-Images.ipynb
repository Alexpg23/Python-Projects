```python
    import imageio
    import matplotlib.pyplot as plt
    
    pic = imageio.imread('shark.jpg')
    plt.figure(figsize = (10,10))
    plt.imshow(pic)
```

```python
# pic is numpy array with 3 channels in RGB order
pic_red = pic.copy()  # Duplicate image
pic_red[:, :, 1] = 0    # Zero out contribution from green
pic_red[:, :, 2] = 0    # Zero out contribution from blue

plt.figure(figsize = (10,10))
plt.axis('off') 
plt.imshow(pic_red)
```
