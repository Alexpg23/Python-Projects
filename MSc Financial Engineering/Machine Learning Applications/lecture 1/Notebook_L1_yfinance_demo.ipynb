## How to work with a Ticker object

```python
import yfinance as yf
symbol = 'AAPL'
ticker = yf.Ticker(symbol)
```

```python
pd.Series(ticker.info).head(20)
```

### Get market data

```python
data = ticker.history(period='5d',
                      interval='1m',
                      start=None,
                      end=None,
                      actions=True,
                      auto_adjust=True,
                      back_adjust=False)
data.info()
```
