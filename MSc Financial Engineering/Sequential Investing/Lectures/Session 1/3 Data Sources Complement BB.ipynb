## The resample() method

Let's download data for the FTSE 100 index:

```python
ftse = yf.download("^FTSE", start = "2000-01-04", end = "2019-10-11")
```

Note that the values of the "Date" column are already defined as an index of the type DateTime, so we don't have to do it manually:

```python
ftse.info()
```

```python
ftse.head()
```

We can downsample the data into weekly ("W") bins and select the last values of the timestamps falling into a bin.

```python
ftse.resample("W").last()
```

Notice that the default day is Sunday, as demonstrated below through the use of a simple function:

```python
import datetime
import calendar
```

```python
def find_day(date):
    input_date = datetime.datetime.strptime(date, "%Y-%m-%d").weekday()
    return calendar.day_name[input_date]
```

```python
find_day("2000-01-09")
```

```python
find_day("2000-01-16")
```

```python
find_day("2000-01-23")
```

```python
find_day("2000-01-30")
```

```python
find_day("2000-02-06")
```

Therefore, if you wanted to set a specific day on your weekly bin, for instance Friday, you would write the following code: 

```python
ftse.resample("W-Fri").last()
```

Below we demonstrate that we can also downsample our series into Monthly ("M") bins and sum the values of the timestamps falling into a bin.

```python
ftse.resample("M").sum()
```

We can perform the same downsampling to other fequencuies such as Qyuarterly, Yearly, etc. and use other methods such as mean(), std(), etc.
