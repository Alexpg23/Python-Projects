```python
data['Bid-Ask A Spread']=data['ASK']-data['BID']
data['Bid-Ask R Spread']=data['Bid-Ask A Spread']/((data['ASK']+data['BID'])/2)
data['$_Vol']=np.abs(data['PRC'])*data['VOL']
data['Amihud Illiquidity Ratio']=data['ASK']
#z=data['ewretd'].astype(float)
#z
data
```

```python
ret= [np.nan if x=='.c' else float(x) for x in data['RET']]
data['RET']=ret
data.dropna(inplace=True)
data
```

```python
data['Amihud Illiquidity Ratio']=np.abs(data['RET'])/data['$_Vol']
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.dropna(inplace=True)
data
```

```python
data['Ret for compound']=data['RET']+1
data
```

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
monthly_ret = pd.DataFrame(index=Tickers, columns=months)

for i in Tickers:
    df=data.loc[data['TICKER']==i,'Ret for compound']
    dff=df.groupby(df.index.strftime('%B')).prod()
    dff=dff-1
    dff.index = pd.Categorical(dff.index, categories=months, ordered=True)
    dff.sort_index() 
    monthly_ret.loc[i,:]=dff

monthly_ret
```

```python
print('The average monthly return is: ',monthly_ret.mean().mean())
```

```python
M_Amihud = pd.DataFrame(index=Tickers, columns=months)

for i in Tickers:
    df=data.loc[data['TICKER']==i,'Amihud Illiquidity Ratio']
    dff=df.groupby(df.index.strftime('%B')).mean()
    dff.index = pd.Categorical(dff.index, categories=months, ordered=True)
    dff.sort_index() 
    M_Amihud.loc[i,:]=dff

M_Amihud
```
