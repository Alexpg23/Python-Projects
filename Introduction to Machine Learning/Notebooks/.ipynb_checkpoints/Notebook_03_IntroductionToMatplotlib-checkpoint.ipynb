# Introduction to MatplotLib for Machine Learning
Matplotlib is the basic plotting library that goes with Numpy and Pandas

## Loading and Styles

```python
import matplotlib.pyplot as plt 
# If your plot does not show then you must have forgotten this line
%matplotlib inline
plt.style.use('seaborn')
import numpy as np
```

```python
print(plt.style.available)
```

## Line Plots

```python
x = np.arange(0, 5, 0.1)
y2 = x ** 2
```

```python
x
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y2, 'r')
plt.xlabel('x')
plt.ylabel('x squared')
plt.title('Quadratic Function')
```

```python
y3 = x ** 3
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y2, 'r', label = "Quadratic")
plt.plot(x, y3, 'g', label = "Cubic")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomials')
plt.grid()
plt.legend()
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
cx, sx = np.cos(x), np.sin(x)
plt.figure(figsize=(8,6), dpi=80)
plt.plot(x, cx, color='blue', label='Cos(x)', linewidth=2)
plt.plot(x, sx, color='red', label='Sin(x)', linewidth=2)
plt.title('Trigonometric Functions')
plt.legend()
plt.savefig("Chapter2_SinCosPlot.png")
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.subplot(2, 1, 1)
plt.plot(x, sx)
plt.ylabel('Sin(x)')
plt.title('Trigonometric Functions')
plt.subplot(2, 1, 2)
plt.plot(x, cx)
plt.ylabel('Cos(x)')
plt.xlabel('Degrees (radians)')
plt.savefig("Chapter2_SinCosSubPlots.png")
```

## Histograms


```python
n = np.random.randn(10000)
plt.figure(figsize=(8,6), dpi=80)
plt.hist(n, bins=20, alpha=0.5, density=1)
plt.title('Random Number Frequency Plot')
plt.xlim((min(n), max(n)))
plt.xlabel('X')
plt.ylabel('PDF')
plt.grid(True)
plt.savefig("Chapter2_Histogram.png")
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.hist(n, bins=50, cumulative=True, alpha = 0.5, density=1)
plt.title("Cumulative detailed histogram")
plt.xlim((min(n), max(n)))
```

## Scatterplots

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = x*y*1000  # 0 to 15 point radii

plt.figure(figsize=(12, 6))
plt.style.use('classic')
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```
